```python
def loadCsv(address):
    with open(address) as csvDataFile:
        info=[]
        csvReader = csv.reader(csvDataFile)
        for row in csvReader:
            info.append(row)
    return info[1:]
```

```python
def deleteCol(a,num):
    for row in a :
        del row[num]
    return a
```

```python
def reformatdate(a):
    temp = ''
    for i in a:
        if i=='/':
            i='-'
        temp = temp+i
    return temp
```

```python
for i in sales:
    i.append('0')
    i.append('0')
    
```

```python
map1 = {}
map2 = {}
for i in prom:
    date = reformatdate(i[0])
    if i[1]=='-999':
        map1[(date,i[2])] = i[3]
    else:
        map2[(date,i[1])] = i[3]
```

```python
index=0
num = len(sales)
for i in sales:
    index=index+1
    if index%1000==0:
        print(str(index)+"/"+str(num)+" has been finished")
    
    #date map11
    if (i[2],i[7]) in map1:
        i[8] =1
        i[9] = map1[(i[2],i[7])]
    elif (i[2],i[0]) in map2:
        i[8] = 1
        i[9] = map2[(i[2],i[0])]
```
